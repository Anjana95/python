## Python interpreter

1. For the python to run, python interpreter is required
2. Interpreter is a program that converts high-level python programs into machine codes which the computer understands


1. Source Code Analysis : 
    * Pyton interpreter receives the source code.
    * It checks for any syntax error and if there are any error , stops the program from running
    * Phase is called lexical analysis - divides the source code into list of tokens
2. Byte Code generation : 
    * Parser recieves the list of tokens and manipulates to generate a big structure - AST(Abstract Syntax Tree)
    * The interpreter converts AST to machine code(bytes) which is saved in file .pyc 
3. PVM (Python Virtual Machine) :
    * interpreter initializes the run time engine - PVM
    * Inputs to PVM - machine code with libraries
    * Output from PVM - converts to executable codes(0 and 1's) binary digits
