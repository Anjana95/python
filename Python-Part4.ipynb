### Python Arithmetic Operators Examples

#### Addition operator 

+operator used to add two numbers

```python
a = 1
b = 6

res = a+b
print(res)
```

```python
def add(a,b):
    print(a+b)

add(-9,8)
add(-9,-8)
```

#### Python Arithmetic Operators Examples

'-' operator used to subtract second value from first

```python
def sub(val1,val2):
    print(val1-val2)

sub(1,2)
sub(9,8)
sub(-7,9)
```

#### Multiplication Operator

'*' operator used to find product of two values

```python
def mul(m1,m2):
    print(m1*m2)

mul(6,2)
mul(2,6)
mul(-6,2)
mul(6,-2)
mul(-6,-2)
```

#### Division Operator 

'/' operator used to find the quotient when first operand is divided by second



```python
def div(d1,d2):
    print(d1/d2)

div(6,2)
div(2,6)
div(-6,2)
div(6,-2)
div(-6,-2)
```

#### Modulus Operator

'%' Modulus operator is used to find the remainder when the first operand is divided by the second


```python
def modulus(mod1,mod2):
    print(mod1%mod2)

modulus(6,2)
modulus(2,6)
modulus(-6,2)
modulus(6,-2)
modulus(-6,-2)
```

#### Exponentiation Operator

'**' used to raise the first operand to the power of second

```python
def exp(e1,e2):
    print(e1**e2)

exp(6,2)
exp(2,6)
exp(-6,2)
exp(6,-2)
exp(-6,-2)
```

### Floor Division Operator

'//' used to find the floor of the quotient when first operand divided by number 



```python
def floor(f1,f2):
    print(f1//f2)

floor(6,2)
floor(2,6)
floor(-6,2)
floor(6,-2)
floor(-6,-2)
```

## Assignment Operators in Python

1. Operators are used to perform operations on variables and values
2. Special symbols which carry out arithematic, logical and bitwise operations
3. Operand - value the operator works

link : https://www.geeksforgeeks.org/assignment-operators-in-python/?ref=roadmap

#### Assign: 

Assign the right side of the operator to the left side operand

```python
x = 2
y = 3

# Assignment operator
z = x+y

print(z)
```

```python
x+=y
print(x)
```

#### Add and Assign

Used to add the right side operand with the left side operand and assign the result to the left side operand

```python
def add_assign(x,y):
    x+=y
    print(x)

add_assign(8,9)
```

#### Subtract and Assign

Subtract the right operand from the left operand and assigning the value to the left

```python
def sub_operand(a,b):
    a-=b
    print(a)

sub_operand(7,19)
```

#### Multiply and Assign:

Multiply the right operand with the left and assigning the result to the left operand

```python
def mul_operand(x,y):
    x*=y
    print(x)

mul_operand(3,189)
```

#### Divide and Assign

1. Divide the left operand with the right and assign the result to the left 



```python
def div_operand(a,b):
    a/=b
    print(a)

div_operand(4,5)
```

#### Modulus and Assign: 

used to take the modulus using the right and left operands and assign to the left 



```python
def modulus_operand(a,b):
    a%=b
    print(a)

modulus_operand(67,8)
```

#### Divide (floor) and Assign:

Divide the left operand with the right and assign the result to left 


```python
def div_operand(a,b):
    a//=b
    print(a)

div_operand(-10,3)
```

#### Exponent and Assign: 

calculate(raise power) value using operands and assign result to left 

```python
def exp_operand(a,b):
    a**=b
    print(a)
    
exp_operand(2,5)
```

#### Bitwise AND and Assign:

* Used to perform AND on both operands and assign the result to the left



```python
def and_operand(a,b):
    a&=b
    print(a)

and_operand(3,6)
```

#### Bitwise OR and Assign: 

* Used to perform bitwise OR on the operands and assign the result to the left 

```python
def bit_operand(a,b):
    a|=b
    print(a)

bit_operand(3,5)
```

#### Bitwise XOR and Assign:

* Used to perform bitwise XOR on the operands and assign the result to the left 

```python
def xor_operand(a,b):
    a^=b
    print(a)

xor_operand(3,5)
```

#### Bitwise Right Shift and Assign: 
* Used to perform bitwise Right shift on the operands and assign the result to the left 

```python
def right_shift(a,b):
    a>>=b
    print(a)

right_shift(3,5)
```

#### Bitwise Left Shift and Assign: 
* Used to perform bitwise Left shift on the operands and assign the result to the left 

```python
def left_shift(a,b):
    a<<=b
    print(a)

left_shift(3,5)
```

### Relational Operators in Python

1. Used for comparing values
2. Either returns True or False 
3. Also known as Comparison operators

#### 1) Greater than

Returns true if the left operand is greater than right operand. Else False.

```python
a = 8
b = 3

print(a>b)
```

#### 2) Less than

Returns true if the left operand is less than right operand. Else False.


```python
print(a<b)
```

#### 3) Equal to

Returns true if the left operand is equal to right operand. Else False. i.e., both the operands are equal


```python
print(a==b)
```

#### 4) Not equal to

Returns true if the left operand is not equal t right operand.

```python
print(a!=b)
```

#### 5) Greater than equal to

Returns true if the left operand is Greater than equal to right operand. else False

```python
print(a>=b)
```

#### 6) Less than equal to

Returns true if the left operand is less than equal to right operand. else False

```python
print(a<=b)
```
