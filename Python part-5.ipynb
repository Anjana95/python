### Logical AND operator

1. Returns True if both the operands are true else False

```python
a = 9
b = 10
c = -10

if a>0 and b>0:
    print("Both a and b numbers are greater than zero ")
if a>0 and b>0 and c>0:
    print("Both a and b numbers are greater than zero ")
else:
    print("Not all number is greater than zero")
    
```

### Logical not operator

1. Works with single boolean values
2. If the boolean value is True, It returns False else True

```python
a = 20

if not(a%3==0):
    print("20 is divisible by 3")
if not(a%5==0):
    print("20 is divisible by 5")
else:
    print("20 is neither divisible by 5 nor by 3")
```

### Shift Operators

These operators used to shift the bits right and left accordingly 



### Bitwise right shift:

1. shifts the bits to the right and fills 0 in place of void
2. In negative numbers, it fills 0 with 1 



```python
a = 10
print("shifting to one place ",a>>1)
print("shifting to twp places ",a>>2)
print("shifting to twp places ",a>>3)
print("shifting to twp places ",a>>4)
print("shifting to twp places ",a>>13)
```

#### Note : Right shift is equal to floor dividing the number by 2 , 
    
1. if it is a>>1 it's a//2
2. if it is a>>2 a//2*2

```python
a = 10
print("shifting to one place ",a//2)
print("shifting to twp places ",a//4)
print("shifting to twp places ",a//8)
print("shifting to twp places ",a//16)
print("shifting to twp places ",a//(2**13))
```
