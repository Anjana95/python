### Syntax of Python range() function

Syntax: range(start, stop, step)

Parameter :

    start: [ optional ] start value of the sequence
    stop: next value after the end value of the sequence
    step: [ optional ] integer value, denoting the difference between any two numbers in the sequence
Return : Returns an object that represents a sequence of numbers

#### Python range() function takes can be initialized in 3 ways.

    range (stop) takes one argument.
    range (start, stop) takes two arguments.
    range (start, stop, step) takes three arguments.

## break, continue and pass in Python

* Using loops in Python automates and repeats the tasks
* Sometimes, there may arise a condition where you want to exit the loop completely, skip an iteration or ignore that condition
* These can be done by loop control statements

### Break Statement in Python

1. Used to terminate the loop or statement where it ends.
2. After that, control will pass to the next statement, if availaible

#### Syntax of Break Statement

    for/while loop:
        # statement(s)
        if condition:
            break
        # statement(s)
    # end loop
    

### Working on Python Break Statement



```python
string = "Anjana devi"

for i in string:
    if i == "a":
        break
    else:
        print(i)
```

```python
i = 1
while i <10:
    
    #print(f"Before if loop {i}")
    if i%5 == 0:
        i += 1
        #print(f"After if loop {i}")
        break
    else:
        print(i)
        #print(f"In else loop {i}")
        
    i += 1
             
```

#### break statements with nested for loops in Python

```python
for i in range(1,5):
    for j in range(2,6):
        if j%i == 0:
            break
        print(i,j)
```

### Continue Statement in Python

1. continue statement is opposite to that of the break statement
2. Instead of terminating the loop, it forces to next iteration of the loop


Syntax of Continue Statement:
    
        for/while loop:
            #statement(s)
            if condition(s):
                continue
            #Statement
            



```python
i = 1
while i <10:
    
    #print(f"Before if loop {i}")
    if i%5 == 0:
        i += 1
        #print(f"After if loop {i}")
        continue
    else:
        print(i)
        #print(f"In else loop {i}")
        
    i += 1
```
