### Accessing characters in Python String

1. String can be accessed by using the method of Indexing
2. Indexing allows negative address references
3. e.g. -1 refers to the last character, -2 refers to the second last character and so on 
4. While accessing an index out of the range will cause an IndexError. 
5. Only Integers are allowed to be passed as an index, float or other types that will cause a TypeError.



```python
string1 = "Anjana devi"

print(string1[0])
print(string1[-1])
print(string1[3])
print(string1[-5])
```

####  String Slicing

1. This method is used to access a range of characters in the String
2. Slicing in a String is done by using a Slicing operator, i.e., a colon (:)
3.  this method is that the string returned after slicing includes the character at the start index but not the character at the last index.

```python
string2 = "Consistency is the key"

print("Iniitial string is\n",string2)

print(string2[2:6])
print(string2[3:-12])
```

#### Reversing a Python String

```python
#Using indexing

string2[::-1]
```

```python
#Using join and reversed function

string3 = "".join(reversed(string2))
print(string3)
```

### Escape Sequencing in Python

1. While printing Strings with single and double quotes in it causes SyntaxError because String already contains Single and Double Quotes and hence cannot be printed
2. To print them, either triple quotes are used or escape sequence 

```python
str_1 = "I'm "Anjana"."
print(str_1)
```

```python
#Using triple quotes

str_1 = '''I'm "Anjana"'''
print(str_1)
```

```python
#Using excape sequence

str_2 = 'I\'m \"Anjana\"'
print(str_2)
```

```python
#Using tab

str_3 = "Anjana\tDevi"
print(str_3)
```

```python
#Using \n(next line)

str_4 = "Anjana\tDevi\nGireesh\tSakthi"
print(str_4)
```

```python
#Printing paths with escape sequence
path = "c:\\Desktop:\\Anjana"
print(path)
```

#### Ignore escape sequence 

1. To ignore the escape sequences in a String, r or R is used, this implies that the string is a raw string and escape sequences inside it are to be ignored.

```python
str_2 = r'I\'m \"Anjana\"'
print(str_2)
```

```python
path = R"c:\\Desktop:\\Anjana"
print(path)
```

### Formatting of Strings

* Strings in Python can be formatted with the use of format() 
* Format method in String contains curly braces {} as placeholders which can hold arguments according to position or keyword to specify the order.

```python
#Default order
str_0 = "{} {}".format("Data","Scientist")
print("Anjana is a ",str_0)
```

```python
#Positional formatting
str_7 = "{0} {2} {1}".format("hard","women","working")
print("Anjana is a ",str_7)
```

```python
#Keyword formatting
str_9 = "{a}, {b} and {c}".format(a="programming",c="cooking",b="physical pursuits")
print("Anjana\'s interests are ",str_9)
```

#### Format specifiers - binary,exponents etc


```python
#Binary representation of 17
str_1 = "{0:b}".format(17)
print(str_1)
```

```python
bin(17)[2:]
```

```python
#Exponents
str_2 = "{0:e}".format(12.3445)
print(str_2)
```

```python
#Rounding off
str_3 = "{0:.3f}".format(4/7)
print(str_3)
```

#### Format specifiers - string allignment

1. A string can be left, right, or center aligned with the use of format specifiers, separated by a colon(:)
2. The (<) indicates that the string should be aligned to the left, 
    * (>) indicates that the string should be aligned to the right and 
    * (^) indicates that the string should be aligned to the center
3. (<10) means that the string should be aligned to the left within a field of width of 10 characters.

```python
string3 = "|{:<10}|{:^10}|{:>16}|".format("Anjana","Devi","Seetharaman")
print(string3)
```

#### Old-style formatting

* Old-style formatting was done without the use of the format method by using the % operator 

```python
integer = 1.2445566

print("Integer is %.2f"%integer)
```

```python
string = "format"
print("Integer is %s"%string)
```

Parameters:

 Return:  Returns the string in the encoded form

* Python String endswith() method returns True if a string ends with the given suffix, otherwise returns False.

Parameters:

### expandtabs() method in Python

link : https://www.geeksforgeeks.org/python-expandtabs-method/#article-meta-div

Python String expandtabs() Method specifies the amount of space to be substituted with the “\t” symbol in the string.

#### Python String expandtabs() Method Syntax

Syntax : expandtabs(space_size)

Parameters
    tabsize : Specifies the space that is to be replaced with the “\t” symbol in the string. By default the space is 8.

    Return : Returns the modified string with tabs replaced by the space.

```python
str_95 = "\t\tAnjana\t\t"
print(str_95)
print(str_95.count("\t"))
```

```python
str_94 = "\tAnjana\t"
print(str_94)
print("Before using expand tabs with arguements ",str_94.count("\t"))
print("Length before using expand tabs with arguements ",len(str_94))
str_93 = str_94.expandtabs(16)
print(str_93)
print("After using expand tabs with arguements ",str_93.count("\t"))
print("Length After using expand tabs with arguements ",len(str_93))
```

##### TypeError when using expandtabs()

### Python String find() method

link : https://www.geeksforgeeks.org/python-string-find/
        


#### Syntax: str_obj.find(sub, start, end)

Parameters: 

    sub: Substring that needs to be searched in the given string. 
    start (optional): Starting position where the substring needs to be checked within the string. 
    end (optional): End position is the index of the last value for the specified range. It is excluded while checking. 
Return:  Returns the lowest index of the substring if it is found in a given string. If it’s not found then it returns -1.

##### Note:

If the start and end indexes are not provided then by default it takes 0 and length-1 as starting and ending indexes where ending indexes are not included in our search.
The find() method is similar to index(). The only difference is find() returns -1 if the searched string is not found and index() throws an exception in this case.

#### find() With No start and end Argument

```python
words = "Start stop start is an example"
words.lower().find('start')
```

```python
words.lower().find('there')
```

#### find() With start and end Arguments

```python
words = "Start stop start is an example"
words.find('start',2)
```

```python
words.find('start',11,15)
```

#### find() Total Occurrences of a Substring

```python
words.lower().count('start')
```

```python
## Finding the total count without using count
start_index = 0
count = 0
for i in range(len(words)):
    j = words.lower().find('start',start_index)
    if j != -1:
        start_index = j+1
        count += 1
print(count)    
```
