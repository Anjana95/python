#### Excpetion handling

#### What are exceptions handling

1. Exceptions - events that occur when an error or exceptional condition that disrupts the normal flow of the program
2. Exception handling - mechanism to handle run time errors or exceptions that may occur during exceution of program and gracefully recover from those errors


###### Sytax:

try: 

    code that may raise an exception

except ExceptionType1:
  
     code to execute exception type1

except ExceptionType2:
  
    code to execute exception type2

else:(Optional)
  
    code to execute if no exception occured in try block

finally:
  
    code that will always be executed regardeless of exceptions


```python
divident = int(input("Enter the divident: "))
divisor = int(input("Enter the divisor: "))
result = divident/divisor
print("result:",result)
```

```python
try:
    divident = int(input("Enter the divident: "))
    divisor = int(input("Enter the divisor: "))
    result = divident/divisor
    print("result:",result)
except ZeroDivisionError:
    print("Cannot divide by zero")
```

The difference is, In the first statement program throws a run time error 
and it does not handle the same which means it doesn't tell the user the exact error. ZeroDivisionError is not understable but 
when it says in english terms it is understood by the user

```python
try:
    divident = int(input("Enter the divident: "))
    divisor = int(input("Enter the divisor: "))
    result = divident/divisor
    print("result:",result)
except ZeroDivisionError:
    print("Cannot divide by zero")
except ValueError:
    print("Both divident and divisor shoould be a number")
else:
    print("Division performed")
finally:
    print("Execution complete")
    
```

### Python int() Function for Custom Objects

#### int() with __int__() function

```python
class Number:
    value = 9
    
    def __int__(self):
        return self.value

value_of_data = Number()
print(type(value_of_data))
print("number is = {}".format((value_of_data)))
```

1. In the above, we created a class Number with defining a value and a function which returns the value
2. Created "value_of_data" object of the Number class which invokes __int__() method 
3. Object stores the value which Number class returns and is printed using int function

```python
class Number():
    value = 100
    
    def __index__(self):
        return self.value
    
    

data_obj = Number()
print("Data obj number is {}".format(int(data_obj)))
```

The int() function in Python is used in all the standard conversions. For example, the conversion of binary to decimal, octal to decimal, and hexadecimal to decimal.

### __ init __

1. init method is invoked without any call 

```python
class Char_string():
    
    def __init__(self,arg):
        self.arg = arg
        

if __name__ == '__main__':
    string1 = Char_string('Anjana')
        
    print(string1)
```

1. Above returns the address of the string , __repr__ used to return the string

```python
class Char_string():
    
    def __init__(self,arg):
        self.arg = arg
        
    def __repr__(self):
        return 'Word is {}'.format(self.arg)

if __name__ == '__main__':
    string1 = Char_string("Anjana")
    print(string1)
    
```

```python
class Char_string():
    
    def __init__(self,arg):
        self.arg = arg
        
    def __repr__(self):
        return 'Word is {}'.format(self.arg)

if __name__ == '__main__':
    string1 = Char_string("Anjana")
    string2 = string1 + " Devi"
    print(string2)
```

```python
class Char_string():
    
    def __init__(self,arg):
        self.arg = arg
        
    def __repr__(self):
        return 'Word is {}'.format(self.arg)
    
    def __add__(self,other):
        return self.arg+other
        

if __name__ == '__main__':
    string1 = Char_string("Anjana")
    string2 = string1 + " Devi"
    print(string2)
```
